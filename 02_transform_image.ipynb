```python
import torch
import torchvision
from torchvision.io import read_image
import torchvision.transforms as T
  

random_picture = read_image('Chelsea.png')
  
# convert this torch tensor to PIL (Python Image Library which deals with images) image. it's the original library that enabled Python to deal with images.

before_transform = T.ToPILImage()(random_picture) # ToPILImage also one transform to convert into PIL image
  
# display my original image

# before_transform.show()

preprocess = T.Compose([
   T.RandomHorizontalFlip(),             # flip the image horizontally
   # T.Resize(50),                       - resize my original image
   # T.CenterCrop(200)                   - crops 200 size of center from my original image
   # T.ToTensor(),                       - converts PIL image -> tensor value (0~1)
   # T.ColorJitter(20,30)                - Randomly change the brightness, contrast, saturation and hue of an image.
   T.Grayscale()                         # Converts my image into grayscale image
   # T.FiveCrop(10)                      - Crop the given image into four corners and the central crop
   # T.RandomVerticalFlip()              - flip vertically
   
]) # we use Compose to do several transforms together

after_transform = preprocess(before_transform)

after_transform
```
